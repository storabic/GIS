```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
plt.imshow(mpimg.imread('girlface.bmp'), cmap="gray")
```

```python
from skimage.io import imread, imsave
from PIL import Image
img = Image.open('girlface.bmp')
```

```python
img
```

```python
arr = np.array(list(img.getdata(band=0)), float)
arr.shape = (img.size[1], img.size[0])
imageee = Image.fromarray(arr).convert('L').save("source.png")
print(arr.shape[0])
plt.imshow(arr, cmap="gray")
```

```python
ggx = np.array([[0,0,0,5,0,0,0],[0,5,18,32,18,5,0],[0,18,64,100,64,18,0],[5,32,100,100,100,32,5],[0,18,64,100,64,18,0],[0,5,18,32,18,5,0],[0,0,0,5,0,0,0]])

ggy = ggx[:]

gx = np.zeros((arr.shape[0] - ggx.shape[0] + 1, arr.shape[1] - ggx.shape[0] + 1))
print(len(gx))
gy = gx[:]
```

```python
def simpleMult(mat, filt):
    return np.sum(np.multiply(mat, filt))
```

```python
for i in range(len(gx)):
    for j in range(len(gx)):
        gx[i, j] = simpleMult(arr[i:i+7, j:j+7], ggx)
        gy[i, j] = simpleMult(arr[i:i+7, j:j+7], ggy)

ans = np.zeros((gx.shape[0]//2, gx.shape[0]//2))
print(ans.shape)
for i in range(len(ans)):
    for j in range(len(ans)):
        #ans[i][j] = (gx[i,j]**2 + gy[i,j]**2)**0.5
        ans[i][j] = (gx[i*2,j*2]+gx[i*2+1, j*2] + gx[i*2+1, j*2+1] + gx[i*2,j*2+1])/4
print(ans.shape)
```

```python
plt.imshow(ans, cmap="gray")
```

```python
ans = ans / 1000
Image.fromarray(ans).convert('L').save("blurGaus3.png")
```
