```python
import numpy as np
```

```python
from PIL import Image
img=Image.open('girlface.bmp')
print(img)
```

```python
import matplotlib.pyplot as plt
imgray=img.copy()
arr=np.array(list(imgray.getdata(band=0)), float)
arr.shape=(imgray.size[1],imgray.size[0])

plt.imshow(arr,cmap="gray")
toSave1 = Image.fromarray(arr)
toSave1.convert('L').save("outputBase.png")
```

```python
h= np.array([[1/8,1/8,1/4,0,1/2,0,0,0],
    [1/8,1/8,1/4,0,-1/2,0,0,0],
    [1/8,1/8,-1/4,0,0,1/2,0,0],
    [1/8,1/8,-1/4,0,0,-1/2,0,0],
    [1/8,-1/8,0,1/4,0,0,1/2,0],
    [1/8,-1/8,0,1/4,0,0,-1/2,0],
    [1/8,-1/8,0,-1/4,0,0,0,1/2],
    [1/8,-1/8,0,-1/4,0,0,0,-1/2]])
```

```python
ht = h.transpose()
```

```python
for i in range(0, 512, 8) :
    for j in range(0, 512, 8) : 
        arr[i : i + 8, j : j + 8] = np.matmul(np.matmul(ht,arr[i : i + 8, j : j + 8]), h)
```

```python
plt.imshow(arr,cmap="gray")
```

```python
eps = 99
for i in range(0,512) :
    for j in range(0, 512) :
        if (abs(arr[i][j]) < eps) :
            arr[i][j] = 0
```

```python
for i in range(0, 512, 8) :
    for j in range(0, 512, 8) : 
        arr[i : i + 8, j : j + 8] = np.matmul(np.matmul(np.linalg.inv(ht),arr[i : i + 8, j : j + 8]), np.linalg.inv(h))
```

```python
plt.imshow(arr,cmap="gray")
```

```python
toSave2 = Image.fromarray(arr)
toSave2.convert('L').save("outputCut.png")
```
