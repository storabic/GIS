```python
data = gdal.Open("ColorTif.tif")
imgarr = np.zeros((1000, 1000, 3), 'uint8')
r = data.GetRasterBand(1).ReadAsArray()
g = data.GetRasterBand(2).ReadAsArray()
b = data.GetRasterBand(3).ReadAsArray()
imgarr[...,0] = r
imgarr[...,1] = g
imgarr[...,2] = b
img = Image.fromarray(imgarr)
plt.imshow(img)
```
